# Part 5 Like/Unlike posts

### Liking the top 30 posts of the ‘dilsefoodie'. Print message if you have already liked it.

##### Opening of posts

```python
driver.get('https://www.instagram.com/')
```

```python
search=driver.find_element_by_class_name('TqC_a')
search.click()
search_type=driver.find_element_by_class_name('x3qfX')
search_type.send_keys('dilsefoodie')
time.sleep(3)
dilse=driver.find_element_by_partial_link_text('dilsefoodie')
dilse.click()
```

```python
# Getting URL
# response=requests.get("https://www.instagram.com/dilsefoodie/")
page_data=BeautifulSoup(driver.page_source,'html.parser')
all_post=page_data.find_all(class_="v1Nh3")
```

```python
all_post_url=[]
for i in range(0,30):
    post_url=all_post[i].find('a')
    all_post_url.append("https://www.instagram.com"+post_url['href'])
all_post_url
```

## Like the post

```python

for i in all_post_url:
    driver.get(i)
    time.sleep(3)
    ##check whether already liked or not
    likecheck=BeautifulSoup(driver.page_source,'html.parser')
    z=likecheck.find(class_='_8-yf5')
    y=z['aria-label']
    if y=='Like':
        print('photo already unliked')
    else:
        print('photo already liked')
    like=driver.find_element_by_class_name('fr66n')                                     
    like.click()
    
```

##### Unliking the top 30 posts of the ‘dilsefoodie’. Print message if you have already unliked it.

## Unlike the post

```python
for i in all_post_url:
    driver.get(i)
    time.sleep(3)
    ##check whether already liked or not
    likecheck=BeautifulSoup(driver.page_source,'html.parser')
    z=likecheck.find(class_='_8-yf5')
    y=z['aria-label']
    if y=='Like':
        print('photo already unliked')
    else:
        print('photo already liked')
    
    like=driver.find_element_by_class_name('fr66n')
    like.click()
    
```
